```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit
```

```python
horas = np.array([22, 23, 19, 12, 12, 9, 7, 11, 14])
costo = np.array([23, 25, 20, 20, 20, 15, 14, 14, 16])
plt.scatter(horas, costo)
p = np.polyfit(horas, costo, 1)
print(p)
modelo = np.poly1d(p)
print(modelo)
```
